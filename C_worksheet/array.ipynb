```c
#include <stdio.h>

void print_kor(int value);
float print_average(int score[3]);

int main()
{
    int score[] = {95, 100, 85};
    float average;
    
    print_kor(score[0]);
    
    average = print_average(score);
    
    printf("average score of kor, eng, math is %.2f\n", average);
    
    return 0;
}

void print_kor(int value)
{
    printf("score of korean: %d\n", value);
}

float print_average(int score[3])
{
    int i, sum = 0;
    for(i = 0; i < 3; i++)
    {
        sum += score[i];
    }
    return (float)sum/3;
}
```

```c
// 배열에 있는 1의 개수가 짝수이면 1을 출력하는 프로그램을 작성하시오.

#include <stdio.h>

int main()
{
    int values = {1, 1, 0 , 0, 1, 0, 1};
    int cnt = 0, even = 0;
    
    for(int i = 0; i <= 6; i++)
    {
        if(values[i] == 1)
        {
            cnt++;
        }
    }
    
    if(cnt % 2 == 0)
    {
        even = 1;
    }
    
    printf("%d\n", even);
    
    return 0;
}
```

```c
#include <stdio.h>

int main()
{
    int values[] = {10, 20, 20, 30, 30, 30};
    int su[6];
    int cnt = 0;
    
    su[0] = values[0];
    
    for(int i = 1; i <= 5; i++)
    {
        if(values[i] != values[i-1])
        {
            su[++cnt] = values[i];
        }
    }
    
    for(int i = 0; i <= cnt; i++)
    {
        printf("%d ", su[i]);
    }
    
    return 0;
}
```

```c
#include <stdio.h>

int main()
{
    char word[] = "hello";
    int cnt, top = 0;
    
    for(int i = 0; i <= 4; i++)
    {
        cnt = 0;
        for(int j = i; j <= 4; j++)
        {
            if(word[j] == word[i])
            {
                cnt++;
            }
            if(cnt > top)
            {
                top = cnt;
            }
        }
    }
    printf("%d\n", top);

    return 0;
}
```

```c
// 두 배열의 값을 비교해서 더 큰 값을 저장하는 배열 만들기
#include <stdio.h>

int main()
{
    int value1[] = {2, 4, 8, 10, 11};
    int value2[] = {1, 5, 9, 11, 12};
    int result[5];
    
    for(int i = 0; i <= 4; i++)
    {
        if(value1[i] > value2[i])
            result[i] = value1[i];
        else
            result[i] = value2[i];
    }
    for(int i = 0; i <= 4; i++)
    {
        printf("%d ", result[i]);
    }
}
```
